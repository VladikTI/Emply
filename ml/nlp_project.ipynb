```python
from nltk.tokenize import RegexpTokenizer
nltk.download('stopwords')
from nltk.corpus import stopwords
stop_words = stopwords.words('russian')
filtered_tokens = []
def preprocess1(df):
    df = df.to_string()
    tokenizer = RegexpTokenizer(r'\w+')
    df = tokenizer.tokenize(df)
    df = [token.lower() for token in df]
    for token in df:
        if token not in stop_words:
            filtered_tokens.append(token)
    return filtered_tokens
print(preprocess1(df))
    
```

```python
#Вторая анкета, предобработка
filtered_tokens_2=[]
def preprocess2(data_a1):
    data_a1 = data_a1.to_string()
    tokenizer = RegexpTokenizer(r'\w+')
    data_a1 = tokenizer.tokenize(data_a1)
    data_a1 = [token.lower() for token in data_a1]
    for token in data_a1:
        if token not in stop_words:
            filtered_tokens_2.append(token)
    return filtered_tokens_2
print(preprocess2(data_a1))
```

```python
#TfidfVectorizer
#cosine_similarity
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import linear_kernel

def text_similarity(filtered_tokens, filtered_tokens_2):
    vectorizer = TfidfVectorizer()
    vector1 = vectorizer.fit_transform(filtered_tokens)
    print(np.shape(vector1))
    vector2 = vectorizer.transform(filtered_tokens_2)
    print(np.shape(vector2))
    similarity = linear_kernel(vector1, vector2)

    return similarity
print(text_similarity(filtered_tokens,filtered_tokens_2))
```
