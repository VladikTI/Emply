```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import linear_kernel
def text_similarity(X, y):
    vectorizer = TfidfVectorizer()
    vector1 = vectorizer.fit_transform(X)
    print(vector1.shape)
    vector2 = vectorizer.transform(y)
    print(np.shape(vector2))
    similarity = linear_kernel(vector1, vector2)

    return similarity
print(text_similarity(X,y))
```

```python
#Рекомендательная система
indices = pd.Series(data.index, index = data['Имя'])
print(indices)
vectorizer = TfidfVectorizer()
vector1 = vectorizer.fit_transform(X)
vector2 = vectorizer.transform(y)
similarity = linear_kernel(vector1, vector2)
print(similarity)
def get_recommendations(title, similarity, indices):
    idx = indices[:13]
    # Get the pairwsie similarity scores
    sim_scores = list(enumerate(similarity[idx]))
    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, reverse=True)
    # Get the scores for 10 most similar movies
    sim_scores = sim_scores[1:11]
    # Get the movie indices
    talk_indices = [i[0] for i in sim_scores]
    # Return the top 10 most similar movies
    return data['Имя'].iloc[talk_indices]


# Generate recommendations
print(get_recommendations(y, similarity, indices))
```
