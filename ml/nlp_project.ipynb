```python
import os, json
import pandas as pd
#датасет с анкетами
data = open('forms.json', encoding='utf-8')
data = json.load(data)
data = pd.DataFrame.from_dict(pd.json_normalize(data), orient='columns')
data
```

```python
from nltk.tokenize import word_tokenize
def tokenize_text(text):
    tokens = word_tokenize(text)
    return tokens

data['Имя'] = data['Имя'].apply(tokenize_text)
data['Фамилия'] = data['Фамилия'].apply(tokenize_text)
data['Опыт работы'] = data['Опыт работы'].apply(tokenize_text)
data['Образование'] = data['Образование'].apply(tokenize_text)
data['Проекты'] = data['Проекты'].apply(tokenize_text)
data['Дополнительная информация'] = data['Дополнительная информация'].apply(tokenize_text)
data['Навыки.Язык программирования'] = data['Навыки.Язык программирования'].apply(tokenize_text)
data['Навыки.Фреймворки'] = data['Навыки.Фреймворки'].apply(tokenize_text)
data['Навыки.Базы данных'] = data['Навыки.Базы данных'].apply(tokenize_text)
data['Навыки.Работа с API'] = data['Навыки.Работа с API'].apply(tokenize_text)
data['Навыки.Инструменты'] = data['Навыки.Инструменты'].apply(tokenize_text)
data['Языки.Русский'] = data['Языки.Русский'].apply(tokenize_text)
data['Языки.Английский'] = data['Языки.Английский'].apply(tokenize_text)
```

```python
df = data.copy()
def create_soup(x):
    return ' '.join(x['Опыт работы']) + ' ' + ' '.join(x['Образование']) + ' ' + x['Проекты'] + ' ' + ' '.join(x['Дополнительная информация']
        + ' ' + x['Навыки.Язык программирования'] + ' ' + ' '.join(x['Навыки.Фреймворки']) + ' ' + x['Навыки.Базы данных'] + ' ' + ' '.join(x['Навыки.Работа с API'])
        + ' ' + x['Навыки.Инструменты'] + ' ' + ' '.join(x['Языки.Русский']) + ' ' + x['Языки.Английский'])
df['soup'] = df.apply(create_soup, axis=1)
```

```python
df
```

```python
X = df.loc[0:306]
y = df.iloc[[307]]#target
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
CV = CountVectorizer()
cosine_matrix = CV.fit_transform(X['soup'])
cosine_matrix2 = CV.transform(y['soup'])
cosine_similarity = cosine_similarity(cosine_matrix, cosine_matrix2)
cosine_similarity
```

```python
cosine_similarity = cosine_similarity.flatten()
final_df = pd.DataFrame()
final_df['names'] = X['Имя']
final_df['similarities']=pd.Series(cosine_similarity)
print(final_df.to_markdown())
```

```python
#Рекомендательная система
ans= pd.DataFrame()
indices = pd.Series(X.index, index=X['Имя']).drop_duplicates()
def get_recommendations(y, cosine_sim=cosine_similarity):
    

    sim_scores = list(enumerate(cosine_sim[indices]))
    
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar user forms
    sim_scores = sim_scores[1:11]
    ans = pd.DataFrame(sim_scores, columns=['indexes', 'sim_scores'])
    ans = ans.drop('indexes', axis=1)

    # Get the user indices
    user_indices = [i[0] for i in sim_scores]
    ans['user_indices'] = user_indices
    ans['names'] = X['Имя'].iloc[user_indices].values

    # Return the top 10 most similar forms
    print('Топ 10 кандидатов:')
    return ans
get_recommendations(y, cosine_similarity)
```
